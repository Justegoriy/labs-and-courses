```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Сгенерируем выборку объема 1000 из распределения хи-квадрат с числом степеней свободы равным 5 и построим для нее гистограмму и теоретическую функцию плотности распределения.

```python
dist_chi2 = sts.chi2(df=5)
sample = dist_chi2.rvs(1000)
pdf = dist_chi2.pdf(sample)
plt.figure(figsize=(6,4))
plt.hist(sample, density=True, bins=10)
x = np.linspace(0, 30, 1000)
plt.plot(x, dist_chi2.pdf(x), label="Теоретическая функция\n плотности распределения")
plt.legend()
plt.xlabel("$x$")
plt.ylabel("$probability$")
plt.show
```

Рассчитаем теоретические параметры данного распределения

```python
m = sts.chi2.mean(df=5)
var = sts.chi2.var(df=5)
print(f"Теоретическое матожидание:{m}.\nТеоретическая дисперсия:{var}")
```

Сгенерируем 1000 выборок объема 5 из распределения хи-квадрат с числом степеней свободы равным 5 и построим гистограмму распределения их среднего

```python
k = 5
list_of_means = np.array([])
for _ in range(1000):
    sample = dist_chi2.rvs(k)
    list_of_means = np.append(list_of_means, np.mean(sample))
plt.figure(figsize=(8,6))
plt.hist(list_of_means, density=True, bins=10)
x = np.linspace(0, 10, 1000)
theoretical_norm = sts.norm(m, np.sqrt(var / k)) # генерируем нормальное распределение с параметрами m=5, sigma=sqrt(10 / 5)
plt.plot(x, theoretical_norm.pdf(x), label="Теоретическая функция\n плотности распределения")
plt.legend()
plt.xlabel("$x$")
plt.ylabel("$probability$")
plt.show
```
