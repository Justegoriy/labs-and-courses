```python
import re
import numpy as np
import scipy.spatial.distance
```

# 1 Задание

```python
text = open('data.txt', 'r')
str_list = []
for line in text.readlines():
    temp = re.split('[^a-z]', line.lower().strip())
    str_list.append([i for i in temp if i != ""])
```

```python
word_dict = {}
for i in range(len(str_list)):
    for j in range(len(str_list[i])):
        if str_list[i][j] not in word_dict:
            word_dict[str_list[i][j]] = 1
word_list = list(word_dict.keys())
```

```python
matrix = []
for i in range(len(str_list)):
    matrix.append([])
    for word in word_list:
        matrix[i].append(str_list[i].count(word))
matrix = np.array(matrix)
matrix.shape
```

```python
distances = [1]
for i in range(1, matrix.shape[0]):
    distance = scipy.spatial.distance.cosine(matrix[0,:], matrix[i,:])
    distances.append(distance)
distances
```

```python
import scipy.linalg
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return np.sin(x / 5.0) * np.exp(x / 10.0) + 5 * np.exp(-x / 2.0)
```

```python
A = np.array([[1, 1], [1, 15]])
a =  np.array([1, 15])
b = f(a)
des = scipy.linalg.solve(A, b)
y_pred = np.sum(A * des, axis=1)
x = np.arange(0, 16, 0.1)
y = f(x)
plt.plot(x, y, a, y_pred, 'or')
```

```python
A = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]])
a =  np.array([1, 8, 15])
b = f(a)
des = scipy.linalg.solve(A, b)
y_pred = np.sum(A * des, axis=1)
x = np.arange(0, 16, 0.1)
y = f(x)
plt.plot(x, y, a, y_pred, 'or')
```
