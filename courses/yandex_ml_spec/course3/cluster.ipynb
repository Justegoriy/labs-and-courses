```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('checkins.csv', sep=';')
df
```

```python
df.columns = ['id', 'user_id', 'venue_id', 'latitude', 'longitude', 'created_at']
df = df[(df.latitude != '                   ') & (df.longitude != '                   ')]
df = df.iloc[:-1,:]
df
```

```python
from sklearn.cluster import MeanShift

x = df.iloc[:100000, 3:5]
clustering = MeanShift(bandwidth=0.1).fit(x)
```

```python
pred = clustering.predict(df.iloc[:100000, 3:5])
cluster_num = {}
for el in pred:
    if el not in cluster_num:
        cluster_num[el] = 1
    else:
        cluster_num[el] += 1
best_clusters = []
clusters = clustering.cluster_centers_
for i in range(len(clusters)):
    if cluster_num[i] > 15:
        best_clusters.append(clusters[i])
```

```python
best_clusters = np.array(best_clusters)
best_clusters.shape, clusters.shape
```

```python
offices = np.array([[33.751277, -118.188740],
                    [25.867736, -80.324116],
                    [51.503016, -0.075479],
                    [52.378894, 4.885084],
                    [39.366487, 117.036146],
                    [-33.868457, 151.205134]])
offices
```

```python
distances = []
for el in best_clusters:
    temp = []
    for of in offices:
        temp.append(np.sqrt((el[0] - of[0])**2 + (el[1] - of[1])**2))
    distances.append(min(temp))
 
```
